## Build CNN Model

### Subtask:
Define a simple Convolutional Neural Network (CNN) architecture with `Conv2D`, `MaxPooling2D`, `Flatten`, and `Dense` layers, designed for binary classification with an input shape of 96x96x3, and then print the model summary.


**Reasoning**:
I need to define a Convolutional Neural Network (CNN) architecture as per the subtask's instructions. This involves importing necessary Keras modules, building a Sequential model with specified layers, and then printing the model summary.



```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# 2. Initialize a Sequential model
model = Sequential([
    # 3. Add a Conv2D layer
    Conv2D(32, (3, 3), activation='relu', input_shape=(96, 96, 3)),
    # 4. Add a MaxPooling2D layer
    MaxPooling2D((2, 2)),
    # 5. Add another Conv2D layer
    Conv2D(64, (3, 3), activation='relu'),
    # 6. Add another MaxPooling2D layer
    MaxPooling2D((2, 2)),
    # 7. Flatten the output
    Flatten(),
    # 8. Add a Dense layer with 'relu' activation
    Dense(128, activation='relu'),
    # 9. Add the final Dense output layer with 'sigmoid' activation
    Dense(1, activation='sigmoid')
])

# 10. Print the model summary
print("CNN Model Summary:")
model.summary()

print("CNN model defined and summary printed successfully.")
```

**Reasoning**:
The previous output showed a UserWarning about `input_shape` when using `Conv2D` within a `Sequential` model. To address this, I will explicitly add an `Input` layer at the beginning of the `Sequential` model, which is the recommended practice by Keras.



```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input

# 2. Initialize a Sequential model
model = Sequential([
    # Add an Input layer explicitly as recommended
    Input(shape=(96, 96, 3)),
    # 3. Add a Conv2D layer
    Conv2D(32, (3, 3), activation='relu'),
    # 4. Add a MaxPooling2D layer
    MaxPooling2D((2, 2)),
    # 5. Add another Conv2D layer
    Conv2D(64, (3, 3), activation='relu'),
    # 6. Add another MaxPooling2D layer
    MaxPooling2D((2, 2)),
    # 7. Flatten the output
    Flatten(),
    # 8. Add a Dense layer with 'relu' activation
    Dense(128, activation='relu'),
    # 9. Add the final Dense output layer with 'sigmoid' activation
    Dense(1, activation='sigmoid')
])

# 10. Print the model summary
print("CNN Model Summary:")
model.summary()

print("CNN model defined and summary printed successfully.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model

# 1. Select a random image from the X_val dataset for visualization
np.random.seed(42) # for reproducibility

# Filter X_val and y_val to get images for each label
images_label_0 = X_val[y_val == 0]
images_label_1 = X_val[y_val == 1]

# Select one representative image for each label
# Ensure there are enough images to sample from, otherwise take the first available
sample_image_0 = images_label_0[np.random.randint(0, len(images_label_0))] if len(images_label_0) > 0 else None
sample_image_1 = images_label_1[np.random.randint(0, len(images_label_1))] if len(images_label_1) > 0 else None

print("Displaying one representative image for each label:")

plt.figure(figsize=(8, 4)) # Adjust figure size for two images

# Display image for label 0 (door_closed)
if sample_image_0 is not None:
    plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
    plt.imshow(sample_image_0)
    plt.title("Original Image (Label: 0 - door_closed)")
    plt.axis('off')
else:
    print("No images available for label 0 (door_closed).")

# Display image for label 1 (door_open)
if sample_image_1 is not None:
    plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
    plt.imshow(sample_image_1)
    plt.title("Original Image (Label: 1 - door_open)")
    plt.axis('off')
else:
    print("No images available for label 1 (door_open).")

plt.tight_layout()
plt.show()

# Now, visualize convolutions for a single representative image (e.g., label 0)
# We will use the sample_image_0 for convolution visualization, if available
if sample_image_0 is not None:
    print("\nVisualizing convolutions for a sample 'door_closed' image:")
    input_image_for_model = np.expand_dims(sample_image_0, axis=0)

    if not model.built:
        model.build(input_shape=(None, 96, 96, 3))
        print("Model explicitly built.")

    layer_outputs = [layer.output for layer in model.layers if 'conv2d' in layer.name]
    activation_model = Model(inputs=model.layers[0].input, outputs=layer_outputs)
    feature_maps = activation_model.predict(input_image_for_model)

    layer_names = [layer.name for layer in model.layers if 'conv2d' in layer.name]

    for layer_name, feature_map in zip(layer_names, feature_maps):
        print(f"\nVisualizing feature maps for layer: {layer_name}")

        n_features = feature_map.shape[-1]
        size = feature_map.shape[1]

        display_grid = 4
        rows = int(np.ceil(display_grid / 4))

        plt.figure(figsize=(8, rows * 2))
        for i in range(display_grid):
            if i < n_features:
                ax = plt.subplot(rows, 4, i + 1)
                plt.imshow(feature_map[0, :, :, i], cmap='viridis')
                plt.title(f"Filter {i}")
                plt.axis('off')
            else:
                plt.subplot(rows, 4, i + 1).set_visible(False)
        plt.suptitle(f'Feature Maps for {layer_name}', fontsize=12)
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()
else:
    print("Cannot visualize convolutions as no sample image for label 0 was available.")

print("Feature map visualization complete.")
```

## Convert Model to TensorFlow Lite

### Subtask:
Convert the trained Keras CNN model into a TensorFlow Lite (.tflite) format. This optimized format is crucial for efficient deployment on edge devices and microcontrollers like the SEEED XIAO, minimizing memory footprint and computational requirements. The size of the converted model will be printed.


**Reasoning**:
To convert the Keras model to TensorFlow Lite format as per the instructions, I need to use `tf.lite.TFLiteConverter` to instantiate the converter, then use its `convert()` method, save the result using `tf.io.gfile.GFile`, and finally get the file size using `os.path.getsize`.



```python
import tensorflow as tf
import os

# 1. Instantiate a TFLiteConverter object from the trained Keras model
print("Instantiating TFLiteConverter...")
converter = tf.lite.TFLiteConverter.from_keras_model(model)
print("TFLiteConverter instantiated.")

# 2. Convert the model to TensorFlow Lite format
print("Converting model to TensorFlow Lite format...")
tflite_model = converter.convert()
print("Model conversion complete.")

# 3. Save the converted TensorFlow Lite model to a file
tflite_model_path = 'door_status_model.tflite'
print(f"Saving TensorFlow Lite model to {tflite_model_path}...")
with tf.io.gfile.GFile(tflite_model_path, 'wb') as f:
    f.write(tflite_model)
print("TensorFlow Lite model saved successfully.")

# 4. Get the size of the converted .tflite model in kilobytes and print it
model_size_bytes = os.path.getsize(tflite_model_path)
model_size_kb = model_size_bytes / 1024
print(f"Converted TensorFlow Lite model size: {model_size_kb:.2f} KB")
```
